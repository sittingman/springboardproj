## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info','governor']])
```

#### Top 10 Project Themes

```python
# Take a read of data detail under column 'mjtheme_namecodes'
proj_data_df['mjtheme_namecode'].head(10)
```

```python
# Opening the json file with the json_load function
with open('data/world_bank_projects.json') as f:
    raw = json.load(f)
```

```python
# Convert the 'mjtheme_namecode' string value into a table
proj_theme = json_normalize(raw, 'mjtheme_namecode')
print(proj_theme.head(15))
print(type(proj_theme))
```

```python
# Get the count based occurance of the name
proj_theme['name'].value_counts()
```

```python
# plan on using ffill() to fill missing values. 
# first sort the dataframe by code
proj_theme = proj_theme.sort_values(by=['code']).replace('',np.nan)
# apply ffill()
proj_theme['name'] = proj_theme.groupby(['code'])['name'].ffill()
```

```python
# Check to see if there are still missing value and if all names match the codes
proj_theme.groupby(['code','name']).count()
```

Confirmed that there is no mismatch
